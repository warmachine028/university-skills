Importing the dataSet


```python
# load boston dataset
from sklearn.datasets import load_boston
```

Supress the deprication warning


```python
# to suppress warning
import warnings
with warnings.catch_warnings():  # You should probably not use this dataset.
    warnings.filterwarnings("ignore")
    boston = load_boston()
    x, y = load_boston(return_X_y=True)
print(x.shape)
```

Knowing description of Boston DataSet


```python
print(boston.DESCR)
```

Accessing data of Boston DataFrame


```python
import pandas as pd

data = pd.DataFrame(boston.data, columns=boston.feature_names)
data
```

```python
data["MEDV"] = pd.DataFrame(boston.target)
data
```

- Identifiation of Correlation Matrix of DataFrame
- Round specifies decimal point precision


```python
pd.DataFrame(data.corr().round(2))  # Pearson's Correlation coefficient
```

Deducing the most correlated Attribute to MEDV -> RM


```python
x, y = data["RM"], data["MEDV"]
# X is always the set of features
# y is always the set of target variables
```

The algorithm to use today is Linear Regression


```python
from sklearn.linear_model import LinearRegression

# importing to train our model
from sklearn.model_selection import train_test_split

# returns a model which learns from the data passed to it
lin_reg_classifier = LinearRegression()
```

Converting the data from Series to DataFrame objects


```python
x, y = pd.DataFrame(x), pd.DataFrame(y)
```

Splitting the dataSet using Train split function intpo data and tests

```python
# 20% left for testing 80% left for testing the model
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

To find how many values the training dataSet contains

```python
x_train.shape
```

Now Ultimately Training the Classifier using the data

```python
lin_reg_classifier.fit(x_train, y_train)
```

Now, testing the Prediction of the dataSet

```python
import numpy as np

from sklearn.metrics import mean_squared_error

# list storing the predicted results
y_pred = lin_reg_classifier.predict(x_test)
y_pred
```

Standard Evaluation Techniques to test the Prediction using mean_squared_error.  A Standard Metric

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

Another Standard Metric -> lies between 0 - 1

```python
lin_reg_classifier.score(x_test, y_test)
```

Working with Ridge

```python
from sklearn.linear_model import Ridge
```

Fit the training data and traing label

```python
rig1 = Ridge(alpha=1)
rig1.fit(x_train, y_train)
y_pred = rig1.predict(x_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```
